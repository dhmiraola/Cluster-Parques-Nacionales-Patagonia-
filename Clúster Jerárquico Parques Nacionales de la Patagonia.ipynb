```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_excel("parquesconsolidado1.xlsx")
data.head(10)
```

```python
# se borra la columna parque_nacional que contienen valores int
data = data.drop(['parque_nacional', 'adq_act_actual', 'iniciativas_inversion_actual'], axis=1)
```

## Clustering Jerárquico

```python
# Se crea el clúster jerárquico, en el cuál determino 4 clusters mediante el método 'ward'
clus = AgglomerativeClustering(n_clusters=4, linkage="ward").fit(data)

# se crea una variable md que contenga el grupo de clúster que pertenece cada parque
md = pd.Series(clus.labels_)
```

```python
Z = linkage(data)


plt.figure(figsize=(11,5))
plt.title("Dendrograma")
plt.xlabel("Parque Nacional")
plt.ylabel("Distancia")
dendrogram(Z)
```

```python
# se crea la columna clúster
data["cluster"] = md 

# Nombrar los parques nacionales anteriormente borrados
data["parque_nacional"] = ["Alerce Andino", "Corcovado Norte", "Hornopirén", "Pumalín", "B. O´higgins Norte",
                           "Cerro Castillo", "Corcovado Sur", "Isla Guamblin", "Isla Magdalena", 
                           "Laguna San Rafael", "Melimoyu", "Patagonia", "Queulat", "Agostini + Cabo de Hornos",
                           "B. O´higgins Sur", "Kaweskar", "Pali Aike", "Torres del Paine", "Yendegaia"]

# Función para crear una columna con el nivel de desarrollo según el clúster
def desarrollo(row):
    
    if (row == 0):
        return "medio"
    elif (row == 1):
        return "bajo"
    elif (row == 2):
        return "muy bajo"
    elif (row == 3):
        return "alto"
    else:
        return ""

# Se aplica la función para crear una columna nivel de desarrollo
data["nivel_desarrollo"] = data["cluster"].apply(desarrollo)

data.head(19)
```

```python
subset = data[["cluster","parque_nacional","nivel_desarrollo"]]

subset.head(19)
```
